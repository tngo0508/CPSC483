# 1. Use read_csv() to load and examine the dataset.

# 2. Use sklearn.cluster.KMeans to cluster the dataset using the default parameters. Assign the labels_ attribute determined by the algorithm to a new column in your DataFrame. How many clusters are there?

```python
from sklearn.cluster import KMeans
kmeans = KMeans().fit(df)
df['label'] = kmeans.labels_
df
```

```python
print(set(kmeans.labels_))
```

# 4.Thereâ€™s no particular reason to think that the default value of K is the correct one for this dataset. Letâ€™s switch to hierarchical clustering to see if we can visualize how the data clusters together.
- Plot a scipy.cluster.hierarchy.dendrogram() for the dataset. Single linkage may not be the best choice for the dataset, but it is relatively quick to compute and will help us make a decision on the number of clusters.
- Based on the dendrogram, how many clusters appear to be present in the dataset?


```python
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt
Z = hierarchy.linkage(df, 'single')
plt.figure(figsize=(25,10))
dn = hierarchy.dendrogram(Z)
```

After running many times, the pattern that we can see is that there are 5 clusters

## 6. Now switch to sklearn.cluster.AgglomerativeClustering() and repeat for the same number of clusters. How do the label assignments compare for the two algorithms?
Caution: donâ€™t include the cluster number assigned by K-Means as one of the features to be clustered, or it will skew your results.

```python
from sklearn.cluster import AgglomerativeClustering
df_6 = pd.read_csv('groceries.csv')
clustering = AgglomerativeClustering(n_clusters=5).fit(df_6)
df_6['label'] = clustering.labels_
df_6
```

```python
print(set(clustering.labels_))
```

```python
from collections import Counter
print('Agglomerate clustering\n', Counter(clustering.labels_))
print('K-means\n', Counter(kmeans_n_clusters.labels_))
```

## How do the label assignments compare for the two algorithms?

Even though the labels are different for each cluster, the label assignments for two algorithms are the same. For example, the Aggolomerate clustering, as shown above, has 54 items of label 3. Likewise, K-means has 54 items of label 4. Although 3 is different 4, the number of items for each label are the same for two algorithms.
