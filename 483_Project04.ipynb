## 1. Load and examine the California dataset’s features, target values, and description.

````
from sklearn import datasets
import numpy as np
from sklearn.datasets import fetch_california_housing
import pandas as pd

data_cali = fetch_california_housing()

X, y = data_cali.data, data_cali.target
#print("Dataset's features: ", X)
#print("Dataset's target: ", y)

data_description = data_cali.DESCR
print(data_description)
data_cali.feature_names
X
````

## 2. Recall that when we originally discussed housing prices, we suggested that the price of a house might depend on how many bedrooms it has. Create and fit() an sklearn.linear_model.LinearRegression model using AveBedrms as a predictor of MedHouseVal. How well does the model score()?

````
from sklearn.linear_model import LinearRegression

linReg = LinearRegression()
X_bedrm = (X[:,3]).reshape([-1,1])
y_tg = y.reshape([-1,1])

linReg.fit(X_bedrm, y_tg)
print(f'w0 = {linReg.intercept_}')
print(f'w1 = {linReg.coef_}')

y_pred = linReg.predict(X_bedrm)
Rsquared_score = linReg.score(X_bedrm, y_tg)
print("R squared score: ", Rsquared_score )
````

## 3. Let’s take a closer look at the data. Seaborn’s pairplot() function can be used to plot pairs of features against each other. Plot MedHouseVal as a function of each of the features.


````
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
pd_data_cali = pd.DataFrame({'MedInc': X[:,0],
                             'HouseAge': X[:,1],
                             'AveRooms': X[:,2],
                             'AveBedrms': X[:,3],
                             'Population': X[:,4],
                             'AveOccup': X[:,5],
                             'Latitude' : X[:,6],
                             'Longitude' : X[:,7]})

# Plot between Median House Value and Median Income in block
MedianIncome = pd.DataFrame(pd_data_cali.MedInc)
MedianIncome.insert(1, "target", y_tg)
sns.pairplot(MedianIncome)
print(MedianIncome)
# Plot between Median House Value and median house age in block
HouseAge = pd.DataFrame(pd_data_cali.HouseAge)
HouseAge.insert(1, "target", y_tg)
sns.pairplot(HouseAge)

# Plot between Median House Value and average number of room
AveRooms = pd.DataFrame(pd_data_cali.AveRooms)
AveRooms.insert(1, "target", y_tg)
sns.pairplot(AveRooms)

# Plot between Median House Value and average number of bedrooms
aveBedrms = pd.DataFrame(pd_data_cali.AveBedrms)
aveBedrms.insert(1, "target", y_tg)
sns.pairplot(aveBedrms)

# Plot between Median House Value and block population
population = pd.DataFrame(pd_data_cali.Population)
population.insert(1, "target", y_tg)
sns.pairplot(population)

# Plot between Median House Value and average house occupancy
AveOccup = pd.DataFrame(pd_data_cali.AveOccup)
AveOccup.insert(1, "target", y_tg)
sns.pairplot(AveOccup)

# Plot between Median House Value and house block latitude
Latitude = pd.DataFrame(pd_data_cali.Latitude)
Latitude.insert(1, "target", y_tg)
sns.pairplot(Latitude)

# Plot between Median House Value and house block longitude
Longitude = pd.DataFrame(pd_data_cali.Longitude )
Longitude .insert(1, "target", y_tg)
sns.pairplot(Longitude)
````
